```python
import math

a = float(input("a="))
b = float(input("b="))
c = float(input("c="))

if a == 0:
    if b == 0:
        if c == 0:
            print("Vô số nghiệm")
        else:
            print("Vô nghiệm")
    else:
        x = -c / b
        print("Nghiệm tuyến tính x =", round(x, 6))
else:
    d = b*b - 4*a*c
    if d > 0:
        s = math.sqrt(d)
        x1 = (-b + s) / (2*a)
        x2 = (-b - s) / (2*a)
        print("x1 =", round(x1, 6))
        print("x2 =", round(x2, 6))
    else:
        if d == 0:
            x = -b / (2*a)
            print("Nghiệm kép x =", round(x, 6))
        else:
            print("Không có nghiệm thực")
```
